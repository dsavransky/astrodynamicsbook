# Kepler's Second Law

Kepler's second law reads: "A line segment joining a planet and the Sun sweeps out equal areas in equal time." In our terminology this 'line' is the orbital radius vector.  

<p style="text-align:center;"><img src="img/second_law.png" alt="Kepler's Second Law" width="400" id="fig:second_law"/></p>

Recall that one of the useful relationships we found while writing out the [perifocal frame components](<13-The Perifocal Frame.ipynb#The-Perifocal-Frame>) of the orbital radius and velocity vectors was: 
$$h = r^2\dot\nu \equiv r^2 \frac{\intd{\nu}}{\intd{t}}$$

If we consider the differential area ($\intd{A}$) swept out by the orbital radius in time $\intd{t}$, as illustrated in the <a href="#fig:second_law" >figure</a>, above, we can write:
$$ \intd{A} = \frac{1}{2}r(r\intd{\nu}) =  \frac{1}{2} \underbrace{r^2 \frac{\intd{\nu}}{\intd{t}}}_{\displaystyle h} \intd{t} $$
or:
<div class="alert alert-block alert-info">
    Kepler's second law: 
    $$\frac{\intd{A}}{\intd{t}} = \frac{h}{2}$$
    </div>
 Since $h$ is constant in time, the rate of change in the area swept out by the orbital radius is a constant as well.  The implication here is that the orbital speed must be inversely proportional to the magnitude of the orbital radius.  In order to cover the same amount of area with a small orbital radius as a large orbital radius, the arc traversed in the same unit time in the former case must be much larger than the one in the latter case.  In other words, a body is moving fastest at periapsis and slowest at apoapsis.  For open orbits, as the orbital radius magnitude goes to infinity, the orbital velocity must go to zero (we will confirm this another way in a short while). 

# Kepler's Third Law

"The square of the orbital period is proportional to the cube of the semi-major axis."  Frankly, this one is just mind-blowing.  Do not forget that these laws are based on fitting to observations of planetary orbits as observed from the Earth, *without the aid of a telescope*.  The fact that it is exactly correct is awe-inspiring. And we can almost trivially derive it, benefiting from mathematical tools not at Kepler's disposal.  We simply integrate our second law expression to find:
$$ \int_0^{T_P} \frac{\intd{A}}{\intd{t}} \intd{t} =   \int_0^{T_P} \frac{h}{2} \intd{t} \quad \Longrightarrow \quad A = \frac{h}{2}T_P$$
where we have defined $T_P$ as the **orbital period** - the time it takes $\nu$ to cover the full range of angles from zero to $2\pi$, or, equivalently, the time between two periapsis (or apoapsis) crossings. Recall that, for an ellipse, the total area is given by $A = \pi a b$ and that we have previously found that:
$$\ell = \frac{h^2}{\mu} = \frac{b^2}{a}$$
Putting it all together, we can write:

<div class="alert alert-block alert-info">
    The orbital period:
    $$T_P = \frac{2\pi}{\sqrt{\mu}} a^{\frac{3}{2}}$$
    </div>

Mind. Blown.

A key takeaway here is that the period is dependent only upon the semi-major axis and the gravitational parameter.  Eccentricity does not enter into it, and two orbits with the same semi-major axis will have exactly the same orbital period, regardless of their eccentricities.  This means that bodies on these orbits (if crossing periapsis/apoapsis at the same time) will return to periapsis/apoapsis at the same time, but will be in different spots on their orbits at all other times. 

# Keplerian Orbit Determination

Reviewing everything we have said so far about two-body (or equivalently **Keplerian**) orbits, we note that we can determine absolutely everything about the two-body orbit from a single, *simultaneous* measurement of the orbital radius and velocity vectors (assuming $\mu$ is known).  Together, these two vectors give us the specific angular momentum direction, and all three (along with the gravitational parameter) allow us to solve for the eccentricity vector, completing the definition of the perifocal frame and allowing us to further solve for the semi-major axis and eccentricity.

In the future, we will provide ourselves with many different ways of determining orbit properties from various types of measurements, but this will be our baseline: the orbital radius and velocity vectors measured at the same time will always tell us everything there is to know about a two-body orbit.

# Effective Potential

Speaking of things you can calculate from orbital position and velocity, recall the [specific orbital energy](<11-Gravity.ipynb#Specific-Energy>):
$$\mc E = \frac{v^2}{2} - \frac{\mu}{r}$$

The energy is a constant of motion and therefore provides a constraint between the orbital radius and velocity magnitudes.  Since specific angular momentum is also a constant, and the radius and velocity vectors are orthogonal at the turning point, we can us their magnitudes at periapsis ($h = r_p v_p$) to substitute for the velocity in the energy expression:
$$\mc E = \frac{h^2}{2r_p^2} - \frac{\mu}{r_p}$$
We have shown already that $r_p = a(1-e)$, and that $\ell = a(1-e^2) = \frac{h^2}{\mu}$.  Substituting all of these and simplifying, we find:

<div class="alert alert-block alert-info">
    The specific energy
    $$\mc E = -\frac{\mu}{a}$$
</div>

<div class="alert alert-block alert-warning">
    <b>Exercise</b><br> Verify that this works for both ellipses and hyperbolae. 
</div>

Assuming you've completed the exercise, you should now believe that this expression applies to all orbits (even parabolae, if we recall that the semi-major axis in this case is technically infinite). Closed orbits have negative specific energy.  Open orbits have positive energy, and parabolic orbits have zero specific energy.

Combining these two energy expression together allows us to find:
<div class="alert alert-block alert-info" id="vis-viva">
    The <b>vis-viva equation</b> (or vis-viva integral):
    $$v^2 = \mu \left( \frac{2}{r} - \frac{1}{a}\right)$$
</div>

On a circular orbit, where $r=a$ at all points, the velocity magnitude is constant as well:
$$v_\textrm{circ} = \sqrt{\frac{\mu}{r}}$$

On a parabolic orbit, with $a = \infty$, we have:
$$v_\textrm{parabolic} = \sqrt{2\frac{\mu}{r}} \equiv \sqrt{2}v_\textrm{circ}$$
This value is also known as the **escape velocity**, since it is the minimum velocity magnitude required to change from a closed to an open orbit (and thereby depart (or escape from) your original central body.  

Finally, recall the form of the orbital velocity vector we found by direct differentiating the orbital radius vector written in [perifocal frame components](<13-The Perifocal Frame.ipynb#The-Perifocal-Frame>):
$$  \mf v = \left[ \dot r\cos(\nu) - r\dot\nu\sin(\nu)\right]\hat{\mf e} + \left[\dot r\sin(\nu) + r\dot\nu\cos(\nu)\right]\hat{\mf{q}}$$
Dotting this expression with itself, we find:
$$\mf v \cdot \mf v \equiv v^2 = \dot r^2 + \underbrace{r^2 \dot \nu^2}_{\displaystyle \equiv \frac{h^2}{r^2} }$$

Substituting into our first specific energy expression yields:
$$\mc E = \frac{\dot r^2}{2} \underbrace{- \frac{\mu}{r} +  \frac{h^2}{2r^2}}_{\displaystyle \triangleq U_\textrm{eff}} $$

The **effective potential** ($U_\textrm{eff}$) groups together all terms of the energy expression that are functions of $r$ (rather than its derivatives). 

<div class="alert alert-block alert-danger">
This cannot be said strongly enough: in general, $\dot r \ne \Vert \mf v \Vert$.  The former is the scalar time derivative of the magnitude of the orbital radius vector while the latter is the magnitude of the vector derivative of the orbital radius vector.  There are many cases where one is zero while the other is non-zero (for example at turning points, or everywhere in a circular orbit). Do not conflate the two, and do not confuse them!
    </div>
<div class="alert alert-block alert-warning">
    <b>Exercise</b><br> Prove that, in general, $\dot r \ne \Vert \mf v \Vert$.  Tattoo the proof somewhere you can easily get at for future reference.
</div>

The effective potential thus includes not only the actual potential energy term, but also a part of the kinetic energy term, and so is not a 'true' potential, but it is hugely useful in characterizing two-body orbits and parametrizations of orbital energy as a function of orbital radius magnitude (and ignoring $\dot r$ entirely).   We can find the orbital radius at which the effective potential is zero:
$$- \frac{\mu}{r} +  \frac{h^2}{2r^2} = 0 \quad \Longrightarrow \quad r(U_\textrm{eff}=0) = \frac{h^2}{2\mu}$$

Differentiating with respect to $r$ and setting this equal to zero also allows us to find the minimum $U_\textrm{eff}$:
$$- \frac{h^{2}}{r^{3}} + \frac{\mu}{r^{2}} = 0 \quad \Longrightarrow \quad  r(\min U_\textrm{eff}) = \frac{h^2}{\mu}$$

Comparing these with the conic section expression for orbital radius magnitude:
$$r = \frac{ h^2/\mu}{1 + e\cos\nu}$$
and we see that the $r(\min U_\textrm{eff})$ value corresponds to the case where $e=0$, while the $r(U_\textrm{eff}=0)$ value corresponds to the case where $e=1$ and $\cos\nu = 0$.  That is, these are the turning points of a circular and parabolic orbit, respectively. In fact, that's exactly what $U_\textrm{eff}$ is tracking - the turning points of two-body orbits. 

Let's try to further visualize what's going on.

```python
# We're going to make our first plot! This is hugely exciting
# We'll use the matplotlib package to do our plotting
# and the numpy package for some general array operations
%matplotlib inline
plt.rcParams['figure.dpi'] = 150
plt.rcParams['font.size'] = 20
fig = plt.figure(figsize=(10,5))
ax = fig.add_axes([0,0,1,1])

# Without loss of generality, choose a unit set such that min U_eff = h = 1
Umin = h = 1
mu = np.sqrt(2*h**2*Umin) # this is the graviational parameter
rmin = h**2/mu # this is the radius corresponding to min Ueff
r0 = h**2/2/mu # this is the radius corresponding to 0 Ueff

# Generate curves for Ueff and its two parts
r = np.linspace(r0/10,rmin*10,1000);
ub = h**2/2/r**2
lb = -mu/r
Ueff = ub + lb
yax = [-Umin*2.5,Umin*3]

# Plot and annotate
ax.plot(r,Ueff,'b',r,ub,'r--',r,lb,'g-.',linewidth=3) 
ax.plot([0,max(r)],[0,0],'k--',[0,max(r)],[-Umin,-Umin],
        'k--',[rmin,rmin],yax,'k--',[r0,r0],yax,'k--')
ax.plot(rmin,-Umin,'bo',r0,0,'rd',markersize=15)
ax.set_xlim([0,max(r)])
ax.set_ylim(yax)
ax.set_xticks([0,r0,rmin,max(r)])
ax.set_xticklabels(['$0$','$\\frac{h^2}{2\\mu}$',
                    '$\\frac{h^2}{\\mu}$',
                    '$\\infty\\rightarrow$'])
ax.set_yticks([min(yax),-Umin,0,max(yax)])
ax.set_yticklabels(['$-\\infty\\downarrow$',
                    '$\\min \\, U_\mathrm{eff}$',
                    '$0$','$\\infty \\uparrow$'])
ax.set_xlabel('$r$')
ax.set_ylabel('$U_\\mathrm{eff}$');
```

The solid blue line is $U_\textrm{eff}$ as a function of $r$ for fixed values of $\mu$ and $h$.  The dashed red and dot-dashed green lines are $\frac{h^2}{2r^2}$ and $-\frac{\mu}{r}$ respectively, summing to $U_\textrm{eff}$. As $r$ goes to $\infty$, the former goes to zero and the latter dominates $U_\textrm{eff}$, which asymptotes back to zero.   The red diamond represents $U_\textrm{eff}= 0$ and corresponds to the periapsis radius of a parabolic (escape) orbit. All orbits above this line are open (hyperbolic) and have a single turning point. All orbits lying below the $U_\textrm{eff} = 0$ line are closed and have two turning points (two values of $r$ for each value of $U_\textrm{eff}$ down to the minimum value of $U_\textrm{eff}$, denoted by the blue circle. This represents a circular orbit, which has an infinite number of turning points, but all at the same orbital radius (which is constant throughout the orbit).
