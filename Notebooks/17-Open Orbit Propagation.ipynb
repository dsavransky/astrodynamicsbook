# Open Orbit Propagation
Having established how to find the orbital radius (and therefore components of the orbital radius and velocity vectors in the perifocal frame) for all time for elliptical orbits, we now have the capability to propagate all closed orbits for all time.  As usual, circular orbits are fairly trivial and boring, as all three anomaly angles are equivalent, and the orbital radius magnitude is constant.  

We now turn our attention to open orbits.

# Propagating Parabolic Orbits

Parabolic orbits, like circular ones, can be propagated without numerical iteration. We start with our usual conic section expression, and take $e=1$ to write:
$$r = \frac{\ell}{1 + e\cos(\nu)} =  \frac{\ell}{1 + \cos(\nu)} = \frac{\ell}{2}\left(\tan^2\left(\frac{\nu}{2}\right) + 1 \right)$$
where we have once again used the tangent half-angle identity:
$$\tan\left(\frac{\theta}{2}\right) = \frac{\sin(\theta)}{1 + \cos(\theta)}  \quad \Longrightarrow \quad \tan^2\left(\frac{\theta}{2}\right) = \frac{\sin^2(\theta)}{\left(1 + \cos(\theta)\right)^2} = \frac{1 - \cos^2(\theta)}{\left(1 + \cos(\theta)\right)^2} = \frac{\left(1 + \cos(\theta)\right)\left(1 - \cos(\theta)\right)}{\left(1 + \cos(\theta)\right)^2} = \frac{1 - \cos(\theta)}{1 + \cos(\theta)} $$
Noting that:
$$\frac{1}{2}\left(\frac{1 - \cos(\theta)}{1 + \cos(\theta)} + 1\right) =   \frac{1}{1 + \cos(\theta)} $$
we thus have:
$$\frac{1}{2}\left(\tan^2\left(\frac{\theta}{2}\right) + 1\right) =   \frac{1}{1 + \cos(\theta)}$$

[Recalling](<13-The Perifocal Frame.ipynb#The-Perifocal-Frame>) that $h =  r^2\dot\nu$, and that $h = \sqrt{\mu \ell}$, we can write:
$$\left[\frac{\ell}{2}\left(\tan^2\left(\frac{\nu}{2}\right) + 1 \right)\right]^2 \dot\nu = \sqrt{\mu \ell} \quad \Longrightarrow \quad \left(\tan^2\left(\frac{\nu}{2}\right) + 1 \right)^2\intd{\nu} = 4\sqrt{\frac{\mu}{\ell^3}} \intd{t}$$

Integrating, this becomes:
$$\frac{2}{3} \tan^{3}\left(\frac{\nu}{2}\right) + 2 \tan \left(\frac{\nu}{2}\right) = 4 \sqrt{\frac{\mu}{\ell^3}} +c$$
where $c$ is the constant of integration.  We define:
<div class="alert alert-block alert-info">
    The <b>parabolic anomaly</b>: 
     $$B \triangleq \tan \left(\frac{\nu}{2}\right)$$
    and the <b>parabolic mean motion</b>:
    $$n_p \triangleq 2 \sqrt{\frac{\mu}{\ell^3}}$$
    such that:
    $$M_p \triangleq n_p(t - t_p) = B + \frac{B^3}{3}$$
</div>

where once again, we have remapped our constant of integration to the time of periapse passage.  This form of the time equation is known as **Barker's equation**, named in honor of Thomas Barker an eighteenth century meteorologist and astronomer who made obsessively detailed recordings and tables of such as things as the weather and the positions of objects on parabolic orbits.

<div class="alert alert-block alert-danger">
    <b> Be careful!</b> The thing that we have defined as $M_p$ here is <b>not</b> an angle.  It has the same form as the elliptical mean anomaly (in that it is a constant times a time difference), but it should not, in any way, be interpreted as an angle.  It's range is $(-\infty,\infty)$ and it is in no way periodic. The same is true for the parabolic anomaly.  It is the output of a tangent function, and thus also has a range of all real numbers.  Both of these should be interpreted as unitless values rather than angles.
    </div>

Plugging things back into our original conic section expression, we find:
<div class="alert alert-block alert-info">
    $$r = \frac{\ell}{2}(1  + B^2)$$
    </div>
    
As we have already established that:
$$\frac{1}{2}\left(\tan^2\left(\frac{\nu}{2}\right) + 1\right) =   \frac{1}{1 + \cos(\nu)} \quad\textrm{and}\quad \tan\left(\frac{\nu}{2}\right) = \frac{\sin(\nu)}{1 + \cos(\nu)}$$
we can write:
$$\frac{B}{1+B^2} = \sin\nu$$
meaning that:
    $$\sin(\nu) = \left(\frac{\ell B}{r}\right)$$


Note that this expression only works in the range of $(-\dfrac{\pi}{2}, \dfrac{\pi}{2})$ as this is the range of the arcsine function. Typically, this is sufficient, but if we wish to find the true anomaly in its full formal range of $(-\pi, \pi)$ we also need a cosine term, which we can get by rewriting the original conic section expression as:
   $$ \cos(\nu) = \frac{\ell}{r} - 1$$
    
<div class="alert alert-block alert-danger">
While it is technically true that parabolic orbits can have true anomalies anywhere in the range $(-\pi, \pi)$, as $\vert \nu \vert \to \pi$, its rate of change $\dot\nu \to 0$.  That is, you can only get there in infinite time.  So, in positing this range, we are technically correct (the best kind of correct).
    </div>

For completeness sake, let's also find the time rate of change of $B$:

```python
B,Bd,nu,nud,t = symbols("B, Bdot,nu,nudot,t", real=True)
h,r,l,mu = symbols("h,r,l,mu", real=True, positive=True)
diffmap={B:Bd,nu:nud}
# This is the differential of lB/r - sin(nu)
exp1 = difftotal(B/(1+B**2),t,diffmap) - difftotal(sin(nu),t,diffmap)
Bdexp = sympy.solve(simplify(exp1.subs(nud,h/r**2).subs(cos(nu),l/r - 1).subs(r, l/2*(1+ B**2)).subs(h,sqrt(mu*l))),Bd)
Bdexp[0]
```

So that:
$$\dot B = \sqrt{\frac{\mu}{l^3}}\frac{4}{B^2 + 1}$$

## Perifocal Frame Components as Functions of Parabolic Anomaly

Just as we did with the [perifocal frame components of elliptical orbits](<15-Keplers Time Equation.ipynb#Perifocal-Frame-Components-as-Functions-of-Eccentric-Anomaly>), we can express the orbital radius and velocity vector components in the perifocal frame as functions of $B$.  All we need to do is substitute the various expression above into the component expressions as functions of true anomaly.  Thus:

$$ \mf r = r\cos(\nu) \bhat{e} + r\sin(\nu) \bhat{q} = \frac{\ell}{2}\left(1 - B^2\right) \bhat{e} + lB  \bhat{q}$$

and
$$\mf v = \frac{\mu}{h}\left[ -\sin(\nu) \bhat{e} + (e + \cos(\nu)) \bhat{q}\right] = \sqrt{\frac{\mu}{l}}\left[-\frac{lB}{r} \bhat{e}  +  \left(1 + \frac{l}{r} -1\right) \bhat{q}\right] = \frac{\sqrt{\mu \ell}}{r}\left(-B\bhat{e}  + 1\bhat{q}\right)  $$ 

## Inverting Barker's Equation

As Barker's equation is a cubic polynomial, it has an exact, closed-form solution.

<div class="alert alert-block alert-danger">
Almost all polynomials up through degree 4 have what are known as solutions in radicals (closed-form algebraic solutions), but no polynomials of degree 5 and up have any such solutions.  How weird is that?  See the Abel-Ruffini theorem, and let your trust in basic algebra be shattered forever.
    </div>
    
To solve for the parabolic anomaly given the mean anomaly, we can make use of Cardano's formula (named after general over-achiever Gerolamo Cardano whos was also best buds with Leonardo da Vinci).  For a polynomial of the form: $x^3 + p x + q = 0$, a real root exists if $4p^3 + 27q^2 > 0$. In this case $p = 3$ and $q = -3M_p$ (since we have $B^3 + 3B - 3M_p = 0$) meaning that one real root (and two complex conjugate roots) will always exist.  The real root will have the form:
$$\left( \sqrt{\frac{q^2}{4} + \frac{p^3}{27}} - \frac{q}{2}\right)^\frac{1}{3} + \left( -\sqrt{\frac{q^2}{4} + \frac{p^3}{27}} - \frac{q}{2}\right)^\frac{1}{3}$$

Plugging in the coefficients from Barker's equation, we get:
$$B = \left( \sqrt{\frac{9 M_p^2}{4} + 1} + \frac{3M_p}{2}\right)^\frac{1}{3} + \left( -\sqrt{\frac{9M_p^2}{4} + 1} + \frac{3M_p}{2}\right)^\frac{1}{3}$$

Let's see how well this works:

# Propagating Hyperbolic Orbits

With hyperbolic orbits, we once again return to the land of non-invertible transcendental expressions and the need for numerical integration.  As with the parabolic case, we seek to come up with a set of equations that mirror, as closely as possible, those found for the elliptical case.  

<p style="text-align:center;"><img src="img/hyperbolic_anomaly.png" alt="Hyperbolic Anomaly" width="400" id="fig:hyperbolic_anomaly"/></p>

For hyperbolae, we also introduce an auxiliary quantity - the hyperbolic anomaly ($H$) - but in this case, we define it via an area rather than an angle. Specifically, the shaded area shown in the <a href="#fig:conics2" >figure</a>, which is equal to $\dfrac{a^2H}{2}$. We define an equilateral hyperbola (one with eccentricity equal to $\sqrt{2}$) with the same semi-major axis of the orbit, and whose vertex coincides with the vertex of the hyperbolic orbit (this equilateral hyperbola is analogous to the auxiliary circle used in the elliptical case). The shaded area is bounded on one side by the section of the equilateral hyperbola between the vertex and a point $A$.  $A$ is the point on the equilateral hyperbola that is on the line crossing the current position of the orbiting body on its hyperbolic orbit (P) and orthogonal to the eccentricity vector direction (segment $\overline{CD}$).  The area is bounded on the other sides by the line segment between the vertex and geometric center of both hyperbolae ($C$), which we know to have a magnitude of $\vert a\vert$, and by line segment $\overline{AC}$. 

By the definitions of the hyperbolic functions, we know that the length of $\overline{AD}$ ($A$'s height above the vertex) is equal to $\vert a\vert \sinh(H)$ and that the length of $\overline{AB}$ ($A$'s distance from the line crossing $C$ and orthogonal to the eccentricity vector) is equal to $\vert a\vert \cosh(H)$.  Recalling that $\overline{CF}$ is the linear eccentricity, and thus has a length of $\vert a \vert e$, and making use of the right triangle $\angle FPD$, we can write:

$$ \sinh(H) =  \frac{r\sin\nu}{\vert a \vert \sqrt{e^2 - 1}} \quad \textrm{and} \quad \cosh(H) = \frac{ae + r\cos\nu}{a}$$

This might be a good time to briefly review hyperbolic functions.  The nice thing is, almost everything you know about the trigonometric functions applies to their hyperbolic counterparts.  $\sinh$ is and odd function and $\cosh$ is an even function.   Many of the summation, half-angle, and double-angle expression look nearly the same.  The primary differences usually take the form of flipped signs.  In particular, the analogue to the Pythagorean identity is $\cosh^2{x} - \sinh^2{x} = 1$.

Next, [recalling](<12-Keplers Laws.ipynb#Hyperbolae>) that the conic section solution for hyperbolae has the exact same form as for ellipses:
$$r = \frac{a(1-e^2)}{1 + e\cos\nu}$$
we substitute this into the hyperbolic function expressions and write:
$$ \sinh(H) = \frac{\sin(\nu)\sqrt{e^2 -1}}{1 + e\cos\nu} \quad \textrm{and} \quad  \cosh(H) = \frac{e + \cos\nu}{1+ e\cos(\nu)}$$

Noting that the half-angle hyperbolic tangent relationships are exactly the same as the ones for tangent (except using hyperbolic sine and cosine), we end up with almost exactly the same form of the conversion equation between hyperbolic anomaly and true anomaly as we had for the eccentric anomaly:

<div class="alert alert-block alert-info">
    $$\tanh\left(\frac{H}{2}\right) = \frac{\sqrt{e-1}}{\sqrt{e+1}}\tan\left(\frac{\nu}{2}\right)$$
</div>

## Hyperbolic Time Equation

Our next goal is to derive a form of the time equation for hyperbolic orbits. As with the elliptical case, we need expressions relating the true anomaly to the hyperbolic anomaly.   Taking a step back, we can solve the $\cosh(H)$ and $\sinh(H)$ expressions above for $\cos(\nu)$ and $\sin(\nu)$ as:
$$\cosh(H) = \frac{e + \cos\nu}{1+ e\cos(\nu)} \quad \Longrightarrow \quad \cosh(H) - e = \cos(\nu)\left( 1 - \cos(H) e\right) \quad \Longrightarrow \quad  \cos(\nu) = \frac{\cosh(H) - e}{1 - e\cos(H)}$$

Let's be lazy and let the computer do the other one for us:

```python
e,nu,H = symbols("e,nu,H", real=True)
cosnu = (cosh(H) - e)/(1-e*cosh(H))
sinnu = simplify(sympy.solve(sinh(H) - sin(nu)*sqrt(e**2-1)/(1+e*cos(nu)),sin(nu))[0].subs(cos(nu),cosnu))
sinnu
```

So: 
$$ \sin(\nu) = \frac{\sqrt{e^{2} - 1} \sinh{\left(H \right)}}{e \cosh{\left(H \right)} - 1}$$

At this point, we can proceed with the time equation derivation in exactly the same way that we did it in the elliptical case.  Once again, let's let the computer crunch through this one:

```python
a,nud,Hd,t,r,mu,h = symbols("a,nudot,Hdot,t,r,mu,h", real=True)
diffmap = {nu:nud, H:Hd}
rexp = a*(1-e**2)/(1+e*cos(nu)) #r expression (as opposed to r variable)
ra1 = rexp/a # r/a as function of nu
ra2 = simplify(ra1.subs(cos(nu),cosnu)) # r/a as function of H
dra1 = difftotal(ra1,t,diffmap) #d/dt(r/a(nu))
dra2 = difftotal(ra2,t,diffmap) #d/dt(r/a(H))
# substitute for nu dot and h
dra1 = ((dra1/ra1**2)*r**2/a**2).subs(r**2*nud,h).subs(h,sqrt(mu*a*(1-e**2)))
# and finally subsitute for the sin(nu) expression derived above
dra1 = simplify(dra1.subs(sin(nu),sinnu))
dra1,dra2
```

So what we have (again) is:

$$ \fddt{}\left( \frac{r}{a}\right) = -\dot H e \sinh(H) = -\sqrt{\frac{\mu}{-a^3}}\frac{e  \sinh{\left(H \right)}}{e \cosh{\left(H \right)} -1 } $$

Integrating and re-arranging, we arrive at an equivalent definition of the mean motion and a new form of the time equation:

<div class="alert alert-block alert-info">
    $$n_h \triangleq  \sqrt{\frac{\mu}{-a^3}} $$
    $$M_h \triangleq n_h(t - t_p) = e\sinh(H) - H$$
    </div>

$t_p$ here is again the time of periapse passage, which again comes from the constant of integration.  As with the parabolic case, the hyperbolic anomaly and mean anomaly are non-periodic, and $M_h$ and $H$ (just like $M_p$ and $B$) again flow from $-\infty$ to $\infty$. Unlike the parabolic case, hyperbolic true anomalies are actually limited in their range and will have a maximum (and minimum) possible true anomaly that is a function of the eccentricity.  Specifically, hyperbolic true anomaly will be in the range of:
$$\vert \nu \vert < \pi - \cos^{-1}\left(\frac{1}{e}\right)$$
meaning that the maximum (absolute) value of $\nu$ goes to $90^\circ$ as eccentricity goes to infinity. Note that we will provide a rigorous proof of all of the above when we get to hyperbolic flyby analysis (a bit of a while down the road).

## Newton-Raphson for Hyperbolae

The hyperbolic for of the time equation is again a transcendental function and thus must be numerically inverted to map between time and true anomaly.  We can apply Newton-Raphson iteration in exactly the same way as we did for the elliptical case.  We define our zero-valued function as:
$$f(H) = M_h - (e\sinh(H)  - H )$$
and differentiate with respect to the hyperbolic anomaly:
$$\frac{\intd{f}}{\intd{H}} = 1 - e\cosh(H)$$
Our iterant is therefore given by:
$$H_{n+1} = H_n - \frac{M_h + H_n - e\sinh(H_n)}{1 - e\cosh(H_n)}$$

We again can choose either an arbitrary small value for our convergence tolerance or iterate to machine precision, with the caveat that in this case we cannot simply use the spacing at $2\pi$ as $M_h$ can be arbitrarily large.  If going the machine precision route, the tolerance must be set at runtime based on the given value of $M$.

For initialization, we can again adopt the strategy of using $H_0 = M_h \pm e$, in this case as:
$$H_0 = \begin{cases} M_h+e & 0 \le M_h \le \pi \\ M_h-e   &  \textrm{else} \end{cases}$$

such that all negative values of $M_h$ use an initialization of $M_h-e$.

Alternatively, we note that the hyperbolic equation has essentially the same Taylor expansion as the elliptical version:

```python
sympy.series(e*sinh(H)-H,x=H,x0=0,n=4) # Expanding in E, about 0, to fourth order
```

meaning that we can try the same kind of initialization:
$$E_0 = \begin{cases}
\dfrac{M_h}{e-1} & \dfrac{M_h}{e-1} < \sqrt{\frac{6(e-1)}{e}}\\
\left(\dfrac{6M_h}{e}\right)^{\frac{1}{3}} & \textrm{else}
\end{cases}$$

Let's see how well these work:

```python
M1 = np.linspace(-100,100,200) # arbitrary range of Ms
e1 = np.linspace(1+np.spacing(1),25,100) # By e=25, true anomaly is limited to ~92 degrees
[Ms,es] = np.meshgrid(M1,e1)
Ms = Ms.flatten() #unwrap to 1D array
es = es.flatten()

# Taylor expansion initialization
H0s1 = Ms/(es-1)
mask = es * H0s1**2 > 6 * (es - 1)
H0s1[mask] = np.cbrt(6 * Ms[mask] / es[mask])
H1,c1 = invKepler(Ms,es,E0=H0s1)

# M \pm e initialization
H0s2 = Ms - es
mask = (Ms >= 0) & (Ms <= np.pi)
H0s2[mask] = Ms[mask] + es[mask]
H2,c2 = invKepler(Ms,es,E0=H0s2,maxIter=200) 
np.mean(c1), np.mean(c2)
```

```python
# Plot the results:
fig,ax = plt.subplots(1,2,figsize=(8,3),sharey=True,sharex=True)
fig.subplots_adjust(left=0, bottom=0, right=0.9, top=1, wspace=0.3)
p1 = ax[0].contourf(M1,e1,np.log10(np.reshape(c1,(e1.size,M1.size))),np.log10(np.unique(c1)))
ax[0].set_ylabel("$e$")
ax[0].set_xlabel("$M$ (rad)")
ax[1].set_xlabel("$M$ (rad)")
cb1 = fig.colorbar(p1,ax=ax[0])
ticks = cb1.get_ticks()
cb1.set_ticks(ticks,labels=np.round(10**ticks))
p2 = ax[1].contourf(M1,e1,np.log10(np.reshape(c2,(e1.size,M1.size))),np.log10(np.unique(c2)))
cb2 = fig.colorbar(p2,ax=ax[1])
ticks = cb2.get_ticks()
cb2.set_ticks(ticks,labels=np.round(10**ticks))
```

The Taylor-expansion based initialization continues to prove itself to be remarkably robust and efficient for very large values of both eccentricity and mean anomaly. While the simpler initialization also converges for all of these test points, in this case, the Taylor-expansion is probably the better approach.  For this reason, it is always used by default in our `invKepler` function.

<div class="alert alert-block alert-danger">
As previously stated, there are an enormous number of other initializations that we have not considered here.  These are tabulated in a variety of papers and textbooks.  In particular, Vallado recommends an initialization based on magnitude of eccentricity (with breaks at e = 1.6 and 3.6) which is similarly robust and performant to the Taylor expansion approach shown here.  When solving the time equation for relatively small numbers of data points (for modern computers, 'small' should be interpreted as easily in the hundreds of thousands to millions - the reference invKepler code running on a reasonably nice laptop will crunch through one million data points in less than 1 second).  It is only when you start considering applications where you must invert the time equation many millions of times, or where you are working in a hugely constrained memory/processing environments, that initialization performance starts meriting more careful consideration.
</div>

## Perifocal Frame Components as Functions of Hyperbolic Anomaly

One last thing before we move onward and upward.  We can again express the components of the hyperbolic orbit position and velocity in terms of the hyperbolic anomaly (thereby omitting the need to convert hyperbolic anomalies to true anomalies when propagating these components).  Using our expressions for $\sin(\nu)$ and $\cos(\nu)$ as functions of $H$, and substituting in to the components of $\mf r$, we have:

```python
rexp2 = simplify(rexp.subs(cos(nu),cosnu))
rvec = Matrix([simplify(rexp2*cosnu),simplify(rexp2*sinnu),0])
mat2vec(rvec,['\mathbf{\hat{e}}','\mathbf{\hat{q}}','\mathbf{\hat{h}}'])
```

Recalling that we've also derived:
$$ -\dot H e \sinh(H) = -\sqrt{\frac{\mu}{-a^3}}\frac{e  \sinh{\left(H \right)}}{e \cosh{\left(H \right)} -1 } \quad \Longrightarrow\quad \dot H = n_h\frac{1}{e \cosh{\left(H \right)} -1 } $$

```python
nh = symbols("n_h",real=True)
Hdexp = nh/(e*cosh(H) - 1)
vvec = difftotalmat(rvec,t,diffmap).subs(Hd, Hdexp).subs(simplify(rexp2/a),r/a)
mat2vec(vvec,['\mathbf{\hat{e}}','\mathbf{\hat{q}}','\mathbf{\hat{h}}'])
```

Thus:

$$\mf r = -a\left(\left(e - \cosh(H)\right)\hat{\mf e} + \sqrt{e^2-1}\sinh(H)\hat{\mf{q}}\right) $$
and
$$\mf v = \frac{a^2 n_h}{r} \left(-\sinh(H)\hat{\mf e} + \sqrt{e^2-1}\cosh(H)\hat{\mf{q}}\right)$$

Let's quickly verify:

```python
# set up sample hyperbolic orbit with h = mu = 1 and e = 5
e = 5
h = mu = 1 #arbitrary units
a = h**2/mu/(1 - e**2) #h^2/mu = a(1-e^2)
M = np.linspace(-10,10,50)
H,_,nu = invKepler(M,e,return_nu=True) #invKepler will also calculate nu if you ask it nicely

# compute everything based on true anomaly
r1 = a*(1-e**2)/(1 + e*np.cos(nu))
re1 = r1*np.cos(nu)
rq1 = r1*np.sin(nu)
ve1 = mu/h*(-np.sin(nu))
vq1 = mu/h*(e + np.cos(nu))

# and now based on hyperbolic anomaly
r2 = a*(1 - e*np.cosh(H))
re2 = a*(np.cosh(H) - e)
rq2 = -a*np.sqrt(e**2 - 1)*np.sinh(H)
nh = np.sqrt(mu/-a**3)
ve2 = -a**2*nh/r2*np.sinh(H)
vq2 = a**2*nh*np.sqrt(e**2-1)/r2*np.cosh(H)

# plot results (they should be identical)
fig,ax = plt.subplots(1,2,figsize=(8,3))
fig.subplots_adjust(left=0, bottom=0, right=0.9, top=1, wspace=0.3)
ax[0].plot(re1,rq1,re2,rq2,'.')
ax[0].set_xlabel('$\mathbf{r} \cdot \mathbf{\hat{e}}$')
ax[0].set_ylabel('$\mathbf{r} \cdot \mathbf{\hat{q}}$')

ax[1].plot(ve1,vq1,ve2,vq2,'.')
ax[1].set_xlabel('$\mathbf{v} \cdot \mathbf{\hat{e}}$')
ax[1].set_ylabel('$\mathbf{v} \cdot \mathbf{\hat{q}}$')
```
